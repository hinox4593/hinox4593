```python
import numpy as np
import matplotlib.pyplot as plt


A = 1
f = 5
phi = 0


t = np.linspace(0, 1, 1000)


sine_wave = A * np.sin(2 * np.pi * f * t + phi)


square_wave = A * np.sign(np.sin(2 * np.pi * f * t + phi))


triangle_wave = (2 * A / np.pi) * np.arcsin(
    np.sin(2 * np.pi * f * t + phi)
)


plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, sine_wave)
plt.title("Sine Wave")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, square_wave)
plt.title("Square Wave")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, triangle_wave)
plt.title("Triangle Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```
